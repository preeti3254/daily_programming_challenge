```python
def subarrZ(nums):  
    curr = 0  
    res = []  
    mapp = {0: [-1]}  
    
    for i, num in enumerate(nums):  
        curr += num  
        if curr in mapp:  
            for si in mapp[curr]:  
                res.append((si + 1, i))  
        
        if curr in mapp:  
            mapp[curr].append(i)  
        else:  
            mapp[curr] = [i]  
    return res
print(f"Test case 1: {subarrZ([4, -1, -3, 1, 2, -1])}")
print(f"Test case 2: {subarrZ([1, 2, 3, 4])}")
print(f"Test case 3: {subarrZ( [0, 0, 0])}")
print(f"Test case 4: {subarrZ([-3, 1, 2, -3, 4, 0])}")
print("")
print(f"Test case 5: {subarrZ([1, -1, 2, -2, 3, -3]*10 )}")
```
